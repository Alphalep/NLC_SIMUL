```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

from model import Ising

# Sample data: assume x, y, z define points and u, v, w define vector directions
x, y, z = np.meshgrid(np.linspace(-1, 1, 10), np.linspace(-1, 1, 10), np.linspace(-1, 1, 10))
u, v, w = np.sin(x), np.cos(y), np.exp(z)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Customize quiver options
# - Set length to control rod length
# - Set arrow_length_ratio to 0 for headless rods
# - Set colors for rods and heads separately
quivers = ax.quiver(
    x, y, z, u, v, w,
    length=0.05,
    arrow_length_ratio=0,
    colors=('b'),  # blue rods, red heads
)

# Customize plot view
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Quiver Plot with Rods')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_zlim(-1.2, 1.2)

plt.show()

```
